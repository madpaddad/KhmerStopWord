```python
import pandas as pd
import re

df = pd.read_csv('khmer_pos.csv')
```

```python
df['pos'].unique()
```

```python

def is_valid_pos(pos):
    if not isinstance(pos, str):
        return False

    pos = pos.strip()

    # must end with dot AND only Khmer letters + dots
    return bool(
        pos.endswith('.') and
        re.fullmatch(r'[\u1780-\u17FF.]+', pos)
    )
```

```python
df_clean = df[df['pos'].apply(is_valid_pos)]
```

```python
df = df_clean.query("pos != '.'")
df = df.query("pos != 'អាល.'")
df = df.query("pos != 'អា.'")
df = df.query("pos != 'ច្រ.'")
df = df.query("pos != 'នី.'")
df = df.query("pos != 'សំខ្យា.'")
df = df.query("pos != 'គូ.'")
df = df.query("pos != 'បូ.'")
```

```python
df.to_csv('khmer_pos.csv', index=False)
```
